```python
def copy_files(file_paths, destination_directory):
    # Ensure the destination directory exists, create it if not
    os.makedirs(destination_directory, exist_ok=True)

    for file_path in file_paths:
        if os.path.isfile(file_path):
            # Extract the filename from the file path
            filename = os.path.basename(file_path)
            # Construct the destination path
            destination_path = os.path.join(destination_directory, filename)

            # Copy the file to the destination directory
            shutil.copy2(file_path, destination_path)
            print(f"Copied {filename} to {destination_directory}")

        else:
            print(f"Warning: {file_path} is not a valid file path. Skipping.")

def remove_files(file_paths):
    for file_path in file_paths:
        if os.path.isfile(file_path):
            # Remove the file
            os.remove(file_path)
            print(f"Removed {file_path}")

        else:
            print(f"Warning: {file_path} is not a valid file path. Skipping.")
```

```python
def collect_images_with_prefix(folder_path, prefix):
    image_paths = []

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.startswith(prefix) and file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                image_path = os.path.join(root, file)
                image_paths.append(image_path)

    return image_paths

# Example usage:
folder_path = './archive1/traffic_Data/TEST/'
prefix_to_find = '007'  # Replace with the desired first three digits
matching_images = collect_images_with_prefix(folder_path, prefix_to_find)
print(len(matching_images))
destination_directory = './database2/test/B'

copy_files(matching_images, destination_directory)
```

```python
import os

def rename_images(directory_path, base_name):
    # Get a list of all files in the directory
    files = os.listdir(directory_path)
    
    # Filter only image files (you can customize the extensions if needed)
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]
    
    # Rename the image files
    for i, old_name in enumerate(image_files):
        extension = os.path.splitext(old_name)[1]
        new_name = f"{base_name}_{i:04d}{extension}"
        old_path = os.path.join(directory_path, old_name)
        new_path = os.path.join(directory_path, new_name)
        os.rename(old_path, new_path)
        print(f"Renamed: {old_name} -> {new_name}")
```
